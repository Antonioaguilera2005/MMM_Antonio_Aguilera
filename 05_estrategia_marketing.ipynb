```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import pickle
import os

DATA_PROC = os.path.join('..', 'data', 'processed')
DATA_RAW  = os.path.join('..', 'data', 'raw')

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.figsize'] = (14, 5)
plt.rcParams['axes.spines.top']   = False
plt.rcParams['axes.spines.right'] = False

df = pd.read_csv(os.path.join(DATA_PROC, 'dataset_maestro_semanal.csv'), parse_dates=['fecha'])
inversion_raw = pd.read_csv(os.path.join(DATA_RAW, 'inversion_medios_semanal.csv'), parse_dates=['semana_inicio', 'semana_fin'])
betas_df = pd.read_csv(os.path.join(DATA_PROC, 'betas_bloques.csv'))
base_features = pd.read_csv(os.path.join(DATA_PROC, 'base_features_mmm.csv'))['feature'].tolist()
media_features = pd.read_csv(os.path.join(DATA_PROC, 'media_features_mmm.csv'))['feature'].tolist()

with open(os.path.join(DATA_PROC, 'modelo_mmm_base.pkl'), 'rb') as f:
    model_base = pickle.load(f)
with open(os.path.join(DATA_PROC, 'scaler_mmm_base.pkl'), 'rb') as f:
    scaler_base = pickle.load(f)
with open(os.path.join(DATA_PROC, 'modelo_mmm_media.pkl'), 'rb') as f:
    model_media = pickle.load(f)
with open(os.path.join(DATA_PROC, 'scaler_mmm_media.pkl'), 'rb') as f:
    scaler_media = pickle.load(f)

# Recreamos bloques
df['bloque_digital_perf']      = df['ads_paid_search']  + df['ads_social_paid']
df['bloque_digital_awareness'] = df['ads_display']      + df['ads_video_online']
df['bloque_offline']           = df['ads_exterior']     + df['ads_prensa'] + df['ads_radio_local']
df['bloque_crm']               = df['ads_email_crm']

BLOQUES = ['bloque_digital_perf', 'bloque_digital_awareness', 'bloque_offline', 'bloque_crm']
for b in BLOQUES:
    df[f'x_{b}'] = np.log1p(df[b].clip(lower=0))

# Recreamos controles estructurales usados en el notebook 04
df = df.sort_values('fecha').reset_index(drop=True)
df['tendencia'] = np.arange(len(df))
for yr in [2021, 2022, 2023, 2024]:
    df[f'dummy_{yr}'] = (df['fecha'].dt.year >= yr).astype(int)

BETAS = dict(zip(betas_df['bloque'], betas_df['beta_real']))
MARGEN_BRUTO = 0.55
PRESUPUESTO_ANUAL = 12_000_000

print("✓ Artefactos del MMM cargados")
print("Coeficientes incrementales de bloques (escala log1p):")
for k, v in BETAS.items():
    print(f"  {k:<30} {v:+.4f}")
print("\nBase features esperadas por el modelo:")
for f in base_features:
    print(f"  {f}")
```

```python
def predecir_componentes(df_in):
    Xb = scaler_base.transform(df_in[base_features])
    Xm = scaler_media.transform(df_in[media_features])
    base_pred  = np.maximum(model_base.predict(Xb), 0)
    media_pred = np.maximum(model_media.predict(Xm), 0)
    total_pred = base_pred + media_pred
    return base_pred, media_pred, total_pred

# Predicciones 2024
df_2024 = df[df['fecha'].dt.year == 2024].copy().reset_index(drop=True)
base_pred_2024, media_pred_2024, total_pred_2024 = predecir_componentes(df_2024)

df_2024['pred_base'] = base_pred_2024
df_2024['pred_media'] = media_pred_2024
df_2024['pred_total'] = total_pred_2024

venta_total_2024       = df_2024['venta_neta_eur'].sum()
base_total_2024        = df_2024['pred_base'].sum()
incremental_total_2024 = df_2024['pred_media'].sum()

print("=== DESCOMPOSICIÓN 2024 ===")
print(f"Ventas totales 2024:      {venta_total_2024:>15,.0f} €")
print(f"Base estructural estimada:{base_total_2024:>15,.0f} €  ({base_total_2024/venta_total_2024*100:.1f}%)")
print(f"Incremental marketing:    {incremental_total_2024:>15,.0f} €  ({incremental_total_2024/venta_total_2024*100:.1f}%)")

# Contribución contrafactual por bloque: venta que se pierde si apagamos el bloque
contribuciones = []
for bloque in BLOQUES:
    df_cf = df_2024.copy()
    df_cf[bloque] = 0
    df_cf[f'x_{bloque}'] = 0
    _, _, total_cf = predecir_componentes(df_cf)
    venta_attr = (df_2024['pred_total'].values - total_cf).sum()
    contribuciones.append({'bloque': bloque, 'venta_atribuida': venta_attr})

contrib_df = pd.DataFrame(contribuciones).sort_values('venta_atribuida', ascending=False)
print("\nContribución incremental por bloque (contrafactual):")
display(contrib_df.round(0))
```

```python
# Mapeamos canales del CSV a bloques
CANAL_A_BLOQUE = {
    'Paid Search':   'bloque_digital_perf',
    'Social Paid':   'bloque_digital_perf',
    'Display':       'bloque_digital_awareness',
    'Video Online':  'bloque_digital_awareness',
    'Exterior':      'bloque_offline',
    'Prensa':        'bloque_offline',
    'Radio Local':   'bloque_offline',
    'Email CRM':     'bloque_crm',
}

inv_2024 = inversion_raw[inversion_raw['semana_inicio'].dt.year == 2024].copy()
inv_2024['bloque'] = inv_2024['canal_medio'].map(CANAL_A_BLOQUE)

inv_bloques = (
    inv_2024.groupby('bloque')['inversion_eur']
    .sum().reset_index()
    .sort_values('inversion_eur', ascending=False)
)
inv_bloques['pct'] = (inv_bloques['inversion_eur'] /
                      inv_bloques['inversion_eur'].sum() * 100).round(1)
inv_bloques['coef_positivo'] = inv_bloques['bloque'].map(
    lambda x: BETAS.get(x, 0) > 0
)

total_inv = inv_bloques['inversion_eur'].sum()
print(f"=== INVERSIÓN 2024 POR BLOQUE ===")
print(f"Total: {total_inv:,.0f} €")
display(inv_bloques)

fig, ax = plt.subplots(figsize=(9, 4))
colores = ['steelblue' if r['coef_positivo'] else 'lightcoral'
           for _, r in inv_bloques.iterrows()]
ax.barh(inv_bloques['bloque'], inv_bloques['inversion_eur'], color=colores)
ax.set_title('Inversión 2024 por bloque\n(azul = coeficiente positivo | coral = coeficiente nulo/negativo)')
ax.set_xlabel('Euros (€)')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:,.0f}€'))
plt.tight_layout()
plt.show()
```

```python
# Retorno por bloque usando contribución contrafactual e inversión real
resultados_mroi = []
for _, row in contrib_df.iterrows():
    bloque = row['bloque']
    venta_attr = row['venta_atribuida']
    inv_anual = inv_bloques[inv_bloques['bloque'] == bloque]['inversion_eur'].sum()
    roas_raw = venta_attr / inv_anual if inv_anual > 0 else 0
    roi_raw  = (venta_attr * MARGEN_BRUTO) / inv_anual if inv_anual > 0 else 0

    resultados_mroi.append({
        'bloque':                 bloque,
        'beta_log1p':             BETAS.get(bloque, 0),
        'inversion_2024':         inv_anual,
        'venta_atribuida':        venta_attr,
        'ROAS_ventas_raw':        roas_raw,
        'ROI_margen_bruto_raw':   roi_raw,
        'ROAS_ventas':            round(roas_raw, 4),
        'ROI_margen_bruto':       round(roi_raw, 4),
    })

mroi_df = pd.DataFrame(resultados_mroi).sort_values('ROAS_ventas_raw', ascending=False)
print("=== RETORNO POR BLOQUE ===")
display(mroi_df[['bloque', 'beta_log1p', 'inversion_2024', 'venta_atribuida', 'ROAS_ventas', 'ROI_margen_bruto']].round(4))
```

```python
# Reasignación del pool con retorno observable; si no existe señal suficiente, mantenemos escenario base
umbral_roas = 0.01
inv_pos = mroi_df[mroi_df['ROAS_ventas_raw'] > umbral_roas].copy()
bloques_pos = inv_pos['bloque'].tolist()
inv_pos_total = inv_pos['inversion_2024'].sum()
inv_pos_actual = dict(zip(inv_pos['bloque'], inv_pos['inversion_2024']))

escenarios = {
    'Actual': {b: inv_pos_actual[b] / inv_pos_total for b in bloques_pos} if inv_pos_total > 0 else {},
    'Más Awareness': {'bloque_digital_awareness': 0.55, 'bloque_digital_perf': 0.30, 'bloque_offline': 0.15},
    'Más Performance': {'bloque_digital_awareness': 0.30, 'bloque_digital_perf': 0.55, 'bloque_offline': 0.15},
    'Equilibrado': {'bloque_digital_awareness': 0.40, 'bloque_digital_perf': 0.35, 'bloque_offline': 0.25},
}

resultados_esc = []
if inv_pos_total <= 0 or len(bloques_pos) == 0:
    _, media_base, total_base = predecir_componentes(df_2024)
    esc_df = pd.DataFrame([
        {
            'escenario': 'Actual',
            'venta_incremental_est': round(media_base.sum(), 0),
            'venta_total_est': round(total_base.sum(), 0)
        }
    ])
    print('No hay pool positivo suficiente para recomendar reasignación; se mantiene el escenario base.')
else:
    for nombre, pesos in escenarios.items():
        df_scn = df_2024.copy()
        fila = {'escenario': nombre}

        for bloque in bloques_pos:
            inv_actual = inv_pos_actual.get(bloque, 0)
            inv_obj = inv_pos_total * pesos.get(bloque, 0)
            factor = inv_obj / inv_actual if inv_actual > 0 else 0
            df_scn[bloque] = df_2024[bloque] * factor
            df_scn[f'x_{bloque}'] = np.log1p(df_scn[bloque].clip(lower=0))
            fila[bloque.replace('bloque_', '')] = round(inv_obj, 0)

        _, media_scn, total_scn = predecir_componentes(df_scn)
        fila['venta_incremental_est'] = round(media_scn.sum(), 0)
        fila['venta_total_est'] = round(total_scn.sum(), 0)
        resultados_esc.append(fila)

    esc_df = pd.DataFrame(resultados_esc)

print("=== SIMULADOR — REASIGNACIÓN DEL POOL OBSERVABLE ===")
display(esc_df)

if len(esc_df) > 1:
    base = esc_df.iloc[0]['venta_incremental_est']
    delta = esc_df['venta_incremental_est'] - base
    colores = ['gray'] + ['steelblue' if d > 0 else 'coral' for d in delta[1:]]
else:
    colores = ['gray']

fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(esc_df['escenario'], esc_df['venta_incremental_est'] / 1e6, color=colores)
ax.set_title('Venta incremental estimada por escenario de reasignación')
ax.set_ylabel('Millones €')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:.2f}M€'))
plt.xticks(rotation=10)
plt.tight_layout()
plt.show()
```

```python
roas_awareness = mroi_df[mroi_df['bloque']=='bloque_digital_awareness']['ROAS_ventas'].values[0] if 'bloque_digital_awareness' in mroi_df['bloque'].values else 0
roas_perf      = mroi_df[mroi_df['bloque']=='bloque_digital_perf']['ROAS_ventas'].values[0] if 'bloque_digital_perf' in mroi_df['bloque'].values else 0
roas_offline   = mroi_df[mroi_df['bloque']=='bloque_offline']['ROAS_ventas'].values[0] if 'bloque_offline' in mroi_df['bloque'].values else 0
roi_awareness  = mroi_df[mroi_df['bloque']=='bloque_digital_awareness']['ROI_margen_bruto'].values[0] if 'bloque_digital_awareness' in mroi_df['bloque'].values else 0
esc_mejor      = esc_df.loc[esc_df['venta_incremental_est'].idxmax(), 'escenario']
uplift_mejor   = esc_df['venta_incremental_est'].max() - esc_df.iloc[0]['venta_incremental_est']
roi_opt = (uplift_mejor / inv_pos_total * 100) if inv_pos_total > 0 else 0

print("=" * 65)
print("  ARGUMENTARIO EJECUTIVO — K-MODA MMM 2024")
print("=" * 65)
print(f"""
1. ENFOQUE METODOLÓGICO
   El MMM final se estima en dos etapas:
   • una base estructural sin medios
   • una capa incremental de marketing con coeficientes no negativos
   Además, los bloques de medios se modelan con log1p(adstock) para recoger saturación.

2. MÉTRICAS QUE IMPORTAN
   • la calidad del modelo se evalúa con CV temporal y holdout 2024
   • el R² in-sample solo se usa como referencia descriptiva
   • la atribución por bloque se calcula por contrafactual, no por multiplicación lineal simplificada

3. DESCOMPOSICIÓN 2024
   • Ventas totales:         {venta_total_2024:>15,.0f} €
   • Base estructural:       {base_total_2024:>15,.0f} €
   • Incremental marketing:  {incremental_total_2024:>15,.0f} €
   • % atribuible marketing: {incremental_total_2024/venta_total_2024*100:>14.1f} %

4. LECTURA DE RETORNO
   • Digital Awareness:  ROAS={roas_awareness:.4f}x | ROI bruto={roi_awareness:.4f}x
   • Digital Perf:       ROAS={roas_perf:.4f}x
   • Offline:            ROAS={roas_offline:.4f}x
   La lectura estratégica es más robusta porque el retorno sale del modelo final por apagado contrafactual.

5. RECOMENDACIÓN 2025
   • Escenario óptimo: '{esc_mejor}'
   • Reasignación dentro del pool observable
   • Uplift incremental estimado: +{uplift_mejor:,.0f} €
   • ROI de optimización: {roi_opt:.2f}% sobre el pool observable actual
""")
print("=" * 65)
```
